```python
import random
import base64
import hashlib
from secrets import SystemRandom
from PIL import Image
import io
import os
import numpy as np
import wave
import struct

# Constants
CHUNK_SIZE = 1024 * 1024  # 1MB chunks
BLOCK_SIZE = 16  # AES-standard 16-byte blocks for CBC
KNOWN_ARMSTRONG = [1, 2, 3, 4, 5, 6, 7, 8, 9, 153, 370, 371, 407, 
                  1634, 8208, 9474, 54748, 92727, 93084]

# --- Armstrong Number Functions (Unchanged) ---
def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    power = len(digits)
    return num == sum(d ** power for d in digits)

def generate_random_armstrong_numbers(count=15):
    if count <= len(KNOWN_ARMSTRONG):
        return KNOWN_ARMSTRONG[:count]
    armstrong_numbers = set(KNOWN_ARMSTRONG)
    num = max(KNOWN_ARMSTRONG) + 1
    while len(armstrong_numbers) < count and num <= 10**6:
        if is_armstrong(num):
            armstrong_numbers.add(num)
        num += 1
    return list(armstrong_numbers)[:count]

def generate_key_sequence(length, iv, data):
    armstrong_count = min(50, max(15, length // 1000))
    armstrong_list = generate_random_armstrong_numbers(count=armstrong_count)
    combined_digits = ''.join(str(num) for num in armstrong_list)
    digits = list(combined_digits)
    randomizer = SystemRandom()
    randomizer.shuffle(digits)
    seed_data = data if isinstance(data, bytes) else str(data).encode('utf-8')
    seed = hashlib.sha256(str(iv).encode('utf-8') + seed_data).digest()
    random.seed(seed)
    key_sequence = bytearray()
    for i in range(length):
        randomizer.shuffle(digits)
        window_start = i % max(1, len(digits) - 64)
        key_bits = ''.join(format(int(d), '04b') for d in digits[window_start:window_start+64])
        key_sequence.append(int(key_bits, 2) & 0xFF)
    return key_sequence

# --- CBC Mode Encryption/Decryption Functions ---
def cbc_encrypt(data, iv, key_sequence):
    encrypted = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i in range(0, len(data), BLOCK_SIZE):
        block = data[i:i+BLOCK_SIZE]
        # Pad block if needed
        if len(block) < BLOCK_SIZE:
            block += bytes([BLOCK_SIZE - len(block)] * (BLOCK_SIZE - len(block)))
        
        # XOR with IV or previous cipher block
        xored_block = bytes(b ^ iv_bytes[j % 32] for j, b in enumerate(block))
        
        # Encrypt using key sequence (Armstrong-based)
        encrypted_block = bytes(b ^ key_sequence[i + j] for j, b in enumerate(xored_block))
        encrypted.extend(encrypted_block)
        
        # Update IV for next block (standard CBC)
        iv_bytes = encrypted_block.ljust(32, b'\0')[:32]  # Ensure 32-byte IV
    
    return bytes(encrypted)

def cbc_decrypt(data, iv, key_sequence):
    decrypted = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i in range(0, len(data), BLOCK_SIZE):
        block = data[i:i+BLOCK_SIZE]
        
        # Decrypt using key sequence
        decrypted_block = bytes(b ^ key_sequence[i + j] for j, b in enumerate(block))
        
        # XOR with IV or previous cipher block
        xored_block = bytes(b ^ iv_bytes[j % 32] for j, b in enumerate(decrypted_block))
        decrypted.extend(xored_block)
        
        # Update IV for next block (standard CBC)
        iv_bytes = block.ljust(32, b'\0')[:32]  # Ensure 32-byte IV
    
    # Remove padding if needed
    pad_len = decrypted[-1]
    if pad_len <= BLOCK_SIZE:
        decrypted = decrypted[:-pad_len]
    
    return bytes(decrypted)

# --- Image Encryption/Decryption (CBC Mode) ---
def encrypt_image_pixels(image, iv):
    img_array = np.array(image)
    original_shape = img_array.shape
    original_dtype = img_array.dtype
    
    # Convert to bytes and encrypt
    data_bytes = img_array.tobytes()
    key_sequence = generate_key_sequence(len(data_bytes), iv, data_bytes)
    encrypted_bytes = cbc_encrypt(data_bytes, iv, key_sequence)
    
    # Reconstruct image
    encrypted_array = np.frombuffer(encrypted_bytes, dtype=original_dtype)
    encrypted_array = encrypted_array.reshape(original_shape)
    return Image.fromarray(encrypted_array), iv, key_sequence

def decrypt_image_pixels(encrypted_image, iv, key_sequence):
    img_array = np.array(encrypted_image)
    original_shape = img_array.shape
    original_dtype = img_array.dtype
    
    # Convert to bytes and decrypt
    data_bytes = img_array.tobytes()
    decrypted_bytes = cbc_decrypt(data_bytes, iv, key_sequence)
    
    # Reconstruct image
    decrypted_array = np.frombuffer(decrypted_bytes, dtype=original_dtype)
    decrypted_array = decrypted_array.reshape(original_shape)
    return Image.fromarray(decrypted_array)

# --- Text Encryption/Decryption (CBC Mode) ---
def encrypt_text(text):
    iv = random.getrandbits(256)
    data_bytes = text.encode('utf-8')
    key_sequence = generate_key_sequence(len(data_bytes), iv, data_bytes)
    encrypted_bytes = cbc_encrypt(data_bytes, iv, key_sequence)
    return base64.b64encode(encrypted_bytes).decode('utf-8'), iv, key_sequence

def decrypt_text(encrypted_text, iv, key_sequence):
    encrypted_bytes = base64.b64decode(encrypted_text.encode('utf-8'))
    decrypted_bytes = cbc_decrypt(encrypted_bytes, iv, key_sequence)
    return decrypted_bytes.decode('utf-8')

# --- Audio Encryption/Decryption (CBC Mode) ---
def encrypt_audio(input_path, output_dir=None):
    try:
        with wave.open(input_path, 'rb') as audio_file:
            params = audio_file.getparams()
            frames = audio_file.readframes(params.nframes)
            
            iv = random.getrandbits(256)
            key_sequence = generate_key_sequence(len(frames), iv, frames)
            encrypted_frames = cbc_encrypt(frames, iv, key_sequence)
            
            # Save encrypted audio
            base_name = os.path.basename(input_path)
            name, ext = os.path.splitext(base_name)
            output_dir = output_dir or os.path.dirname(input_path)
            encrypted_path = os.path.join(output_dir, f"{name}_encrypted{ext}")
            
            with wave.open(encrypted_path, 'wb') as out_file:
                out_file.setparams(params)
                out_file.writeframes(encrypted_frames)
            
            return encrypted_path, iv, key_sequence
    except Exception as e:
        print(f"Audio encryption error: {str(e)}")
        return None, None, None

def decrypt_audio(encrypted_path, iv, key_sequence, output_dir=None):
    try:
        with wave.open(encrypted_path, 'rb') as audio_file:
            params = audio_file.getparams()
            frames = audio_file.readframes(params.nframes)
            
            decrypted_frames = cbc_decrypt(frames, iv, key_sequence)
            
            # Save decrypted audio
            base_name = os.path.basename(encrypted_path)
            name, ext = os.path.splitext(base_name)
            name = name.replace('_encrypted', '_decrypted')
            output_dir = output_dir or os.path.dirname(encrypted_path)
            decrypted_path = os.path.join(output_dir, f"{name}{ext}")
            
            with wave.open(decrypted_path, 'wb') as out_file:
                out_file.setparams(params)
                out_file.writeframes(decrypted_frames)
            
            return decrypted_path
    except Exception as e:
        print(f"Audio decryption error: {str(e)}")
        return None

# --- Main Demo Code (Unchanged) ---
if __name__ == "__main__":
    print("Secure Encryption System (CBC Mode)")
    print("=" * 50)
    
    # Text encryption demo
    text = "This is a secret message!"
    print("\nText Encryption Demo:")
    encrypted_text, iv, keys = encrypt_text(text)
    decrypted_text = decrypt_text(encrypted_text, iv, keys)
    print(f"Original: {text}\nDecrypted: {decrypted_text}")
    
    # Image encryption demo
    image_path = "input.jpg"
    output_dir = "output"
    os.makedirs(output_dir, exist_ok=True)
    
    original_img = Image.open(image_path)
    encrypted_img, iv, keys = encrypt_image_pixels(original_img, random.getrandbits(256))
    decrypted_img = decrypt_image_pixels(encrypted_img, iv, keys)
    
    encrypted_img.save(os.path.join(output_dir, "encrypted.jpg"))
    decrypted_img.save(os.path.join(output_dir, "decrypted.jpg"))
    print("\nImage verification:", np.array_equal(np.array(original_img), np.array(decrypted_img)))
    
    # Audio encryption demo
    audio_path = "ainput.wav"
    encrypted_audio, iv, keys = encrypt_audio(audio_path, output_dir)
    decrypted_audio = decrypt_audio(encrypted_audio, iv, keys, output_dir)
    
    with open(audio_path, 'rb') as f1, open(decrypted_audio, 'rb') as f2:
        print("\nAudio verification:", f1.read() == f2.read())
```

```python
import random
import base64
import hashlib
from secrets import SystemRandom
from PIL import Image
import io
import os
import numpy as np
import wave
import struct
import cv2
import pickle
import time

# Constants
CHUNK_SIZE = 1024 * 1024  # 1MB chunks
BLOCK_SIZE = 16  # AES-standard 16-byte blocks for CBC
KNOWN_ARMSTRONG = [1, 2, 3, 4, 5, 6, 7, 8, 9, 153, 370, 371, 407, 
 1634, 8208, 9474, 54748, 92727, 93084, 548834, 
 1741725, 4210818, 9800817, 9926315, 24678050, 
 88593477, 146511208, 472335975, 534494836, 
 912985153, 467930666, 530017410, 130636361, 
 370881626, 568000238, 520302205, 957494104, 
 928255190, 1308306180, 4330504065]

# --- Armstrong Number Functions ---
def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    power = len(digits)
    return num == sum(d ** power for d in digits)

def generate_random_armstrong_numbers(count=15):
    if count <= len(KNOWN_ARMSTRONG):
        return KNOWN_ARMSTRONG[:count]
    armstrong_numbers = set(KNOWN_ARMSTRONG)
    num = max(KNOWN_ARMSTRONG) + 1
    while len(armstrong_numbers) < count and num <= 10**6:
        if is_armstrong(num):
            armstrong_numbers.add(num)
        num += 1
    return list(armstrong_numbers)[:count]

# Updated generate_key_sequence function with limited key size
def generate_key_sequence(length, iv, data, max_key_size=1024*1024):  # 1MB max
    """
    Generates key sequence limited to max_key_size bytes.
    For longer data, the key sequence will be repeated.
    """
    # actual key size needed (minimum of requested length or max)
    key_size = min(length, max_key_size)
    
    armstrong_count = min(50, max(15, key_size // 1000))
    armstrong_list = generate_random_armstrong_numbers(count=armstrong_count)
    combined_digits = ''.join(str(num) for num in armstrong_list)
    digits = list(combined_digits)
    randomizer = SystemRandom()
    randomizer.shuffle(digits)
    seed_data = data if isinstance(data, bytes) else str(data).encode('utf-8')
    seed = hashlib.sha256(str(iv).encode('utf-8') + seed_data).digest()
    random.seed(seed)
    
    key_sequence = bytearray()
    for i in range(key_size):
        randomizer.shuffle(digits)
        window_start = i % max(1, len(digits) - 64)
        key_bits = ''.join(format(int(d), '04b') for d in digits[window_start:window_start+64])
        key_sequence.append(int(key_bits, 2) & 0xFF)
    
    return key_sequence

# CBC encryption/decryption
def cbc_encrypt(data, iv, key_sequence):
    encrypted = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    key_len = len(key_sequence)

    for i in range(0, len(data), BLOCK_SIZE):
        block = data[i:i+BLOCK_SIZE]
        if len(block) < BLOCK_SIZE:
            block += bytes([BLOCK_SIZE - len(block)] * (BLOCK_SIZE - len(block)))
        
        # Use modulo to repeat key sequence if needed
        xored_block = bytes(b ^ iv_bytes[j % 32] for j, b in enumerate(block))
        encrypted_block = bytes(b ^ key_sequence[(i + j) % key_len] for j, b in enumerate(xored_block))
        
        encrypted.extend(encrypted_block)
        iv_bytes = encrypted_block.ljust(32, b'\0')[:32]

    return bytes(encrypted)

def cbc_decrypt(data, iv, key_sequence):
    decrypted = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    key_len = len(key_sequence)

    for i in range(0, len(data), BLOCK_SIZE):
        block = data[i:i+BLOCK_SIZE]
        decrypted_block = bytes(b ^ key_sequence[(i + j) % key_len] for j, b in enumerate(block))
        xored_block = bytes(b ^ iv_bytes[j % 32] for j, b in enumerate(decrypted_block))
        decrypted.extend(xored_block)
        iv_bytes = block.ljust(32, b'\0')[:32]

    pad_len = decrypted[-1]
    if pad_len <= BLOCK_SIZE:
        decrypted = decrypted[:-pad_len]

    return bytes(decrypted)

# --- Image Encryption/Decryption ---
def encrypt_image_pixels(image, iv):
    img_array = np.array(image)
    original_shape = img_array.shape
    original_dtype = img_array.dtype
    data_bytes = img_array.tobytes()
    padded_len = len(data_bytes) + (BLOCK_SIZE - len(data_bytes) % BLOCK_SIZE)
    key_sequence = generate_key_sequence(padded_len, iv, data_bytes)
    encrypted_bytes = cbc_encrypt(data_bytes, iv, key_sequence)
    return encrypted_bytes, original_shape, original_dtype, iv, key_sequence

def decrypt_image_pixels(encrypted_bytes, original_shape, original_dtype, iv, key_sequence):
    decrypted_bytes = cbc_decrypt(encrypted_bytes, iv, key_sequence)
    decrypted_array = np.frombuffer(decrypted_bytes, dtype=original_dtype)
    decrypted_array = decrypted_array.reshape(original_shape)
    return Image.fromarray(decrypted_array)


if __name__ == "__main__":
    print("Secure Encryption System (CBC Mode)")
    print("=" * 50)

    image_path = "input.jpg"
    output_dir = "output"
    os.makedirs(output_dir, exist_ok=True)

    original_img = Image.open(image_path)
    encrypted_bytes, shape, dtype, iv, keys = encrypt_image_pixels(original_img, random.getrandbits(256))
    decrypted_img = decrypt_image_pixels(encrypted_bytes, shape, dtype, iv, keys)

    # Save encrypted image using raw encrypted bytes (converted to array and reshaped for visualization only)
    encrypted_array = np.frombuffer(encrypted_bytes, dtype=dtype)[:np.prod(shape)].reshape(shape)
    Image.fromarray(encrypted_array).save(os.path.join(output_dir, "encrypted.jpg"))
    decrypted_img.save(os.path.join(output_dir, "decrypted.jpg"))
    print("\nImage verification:", np.array_equal(np.array(original_img), np.array(decrypted_img)))


```
