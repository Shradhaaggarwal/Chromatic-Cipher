```python
import random
import base64
import hashlib
from secrets import SystemRandom
from PIL import Image
import io
import os
import numpy as np

# Constants
CHUNK_SIZE = 1024 * 1024  # 1MB chunks
KNOWN_ARMSTRONG = [1, 2, 3, 4, 5, 6, 7, 8, 9, 153, 370, 371, 407, 
                  1634, 8208, 9474, 54748, 92727, 93084]

def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    power = len(digits)
    return num == sum(d ** power for d in digits)

def generate_random_armstrong_numbers(count=15):
    if count <= len(KNOWN_ARMSTRONG):
        return KNOWN_ARMSTRONG[:count]
    
    armstrong_numbers = set(KNOWN_ARMSTRONG)
    num = max(KNOWN_ARMSTRONG) + 1
    
    while len(armstrong_numbers) < count and num <= 10**6:
        if is_armstrong(num):
            armstrong_numbers.add(num)
        num += 1
    
    return list(armstrong_numbers)[:count]

def generate_key_sequence(length, iv, data):
    armstrong_count = min(50, max(15, length // 1000))
    armstrong_list = generate_random_armstrong_numbers(count=armstrong_count)
    combined_digits = ''.join(str(num) for num in armstrong_list)
    digits = list(combined_digits)

    randomizer = SystemRandom()
    randomizer.shuffle(digits)

    seed_data = data if isinstance(data, bytes) else str(data).encode('utf-8')
    seed = hashlib.sha256(str(iv).encode('utf-8') + seed_data).digest()
    random.seed(seed)

    key_sequence = bytearray()
    for i in range(length):
        randomizer.shuffle(digits)
        window_start = i % max(1, len(digits) - 64)
        key_bits = ''.join(format(int(d), '04b') for d in digits[window_start:window_start+64])
        key_sequence.append(int(key_bits, 2) & 0xFF)

    return key_sequence

def encrypt_image_pixels(image, iv):
    """Encrypt only the pixel data while preserving image structure"""
    img_array = np.array(image)
    original_shape = img_array.shape
    
    flat_array = img_array.flatten()
    data_bytes = flat_array.tobytes()
    
    key_sequence = generate_key_sequence(len(data_bytes), iv, data_bytes)
    
    encrypted_bytes = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i, byte in enumerate(data_bytes):
        iv_segment = int.from_bytes(iv_bytes[i % 25 : (i % 25) + 8], 'big')
        encrypted_byte = byte ^ key_sequence[i] ^ (iv_segment & 0xFF)
        encrypted_bytes.append(encrypted_byte)
        
        iv = int.from_bytes(iv_bytes, 'big') ^ encrypted_byte
        iv_bytes = iv.to_bytes(32, 'big')
    
    encrypted_array = np.frombuffer(bytes(encrypted_bytes), dtype=img_array.dtype)
    encrypted_array = encrypted_array.reshape(original_shape)
    
    return Image.fromarray(encrypted_array), iv, key_sequence

def decrypt_image_pixels(encrypted_image, iv, key_sequence):
    """Decrypt the pixel data"""
    img_array = np.array(encrypted_image)
    original_shape = img_array.shape
    
    flat_array = img_array.flatten()
    data_bytes = flat_array.tobytes()
    
    decrypted_bytes = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i, byte in enumerate(data_bytes):
        iv_segment = int.from_bytes(iv_bytes[i % 25 : (i % 25) + 8], 'big')
        decrypted_byte = byte ^ key_sequence[i] ^ (iv_segment & 0xFF)
        decrypted_bytes.append(decrypted_byte)
        
        iv = int.from_bytes(iv_bytes, 'big') ^ byte
        iv_bytes = iv.to_bytes(32, 'big')
    
    decrypted_array = np.frombuffer(bytes(decrypted_bytes), dtype=img_array.dtype)
    decrypted_array = decrypted_array.reshape(original_shape)
    
    return Image.fromarray(decrypted_array)

def encrypt_text(text):
    """Encrypt text using Armstrong number algorithm"""
    iv = random.getrandbits(256)
    data_bytes = text.encode('utf-8')
    
    key_sequence = generate_key_sequence(len(data_bytes), iv, data_bytes)
    
    encrypted_bytes = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i, byte in enumerate(data_bytes):
        iv_segment = int.from_bytes(iv_bytes[i % 25 : (i % 25) + 8], 'big')
        encrypted_byte = byte ^ key_sequence[i] ^ (iv_segment & 0xFF)
        encrypted_bytes.append(encrypted_byte)
        
        iv = int.from_bytes(iv_bytes, 'big') ^ encrypted_byte
        iv_bytes = iv.to_bytes(32, 'big')
    
    return base64.b64encode(bytes(encrypted_bytes)).decode('utf-8'), iv, key_sequence

def decrypt_text(encrypted_text, iv, key_sequence):
    """Decrypt text using Armstrong number algorithm"""
    encrypted_bytes = base64.b64decode(encrypted_text.encode('utf-8'))
    
    decrypted_bytes = bytearray()
    iv_bytes = iv.to_bytes(32, 'big')
    
    for i, byte in enumerate(encrypted_bytes):
        iv_segment = int.from_bytes(iv_bytes[i % 25 : (i % 25) + 8], 'big')
        decrypted_byte = byte ^ key_sequence[i] ^ (iv_segment & 0xFF)
        decrypted_bytes.append(decrypted_byte)
        
        iv = int.from_bytes(iv_bytes, 'big') ^ byte
        iv_bytes = iv.to_bytes(32, 'big')
    
    return bytes(decrypted_bytes).decode('utf-8', errors='replace')

def process_image(input_path, output_dir=None):
    try:
        if not os.path.exists(input_path):
            raise FileNotFoundError(f"Image not found: {input_path}")

        with Image.open(input_path) as img:
            print(f"\nProcessing: {input_path}")
            print(f"Original Format: {img.format}, Size: {img.size}, Mode: {img.mode}")

            if img.mode != 'RGB':
                img = img.convert('RGB')

            iv = random.getrandbits(256)
            encrypted_img, iv, keys = encrypt_image_pixels(img, iv)
            
            base_name = os.path.basename(input_path)
            name, ext = os.path.splitext(base_name)
            output_dir = output_dir or os.path.dirname(input_path)
            
            encrypted_path = os.path.join(output_dir, f"{name}_encrypted.jpg")
            encrypted_img.save(encrypted_path, format='JPEG', quality=95)
            print(f"Viewable encrypted JPG saved to: {encrypted_path}")

            decrypted_img = decrypt_image_pixels(encrypted_img, iv, keys)
            
            decrypted_path = os.path.join(output_dir, f"{name}_decrypted.jpg")
            decrypted_img.save(decrypted_path, format='JPEG', quality=95)
            print(f"Decrypted JPG saved to: {decrypted_path}")
            
            return encrypted_path, decrypted_path

    except Exception as e:
        print(f"\nError processing {input_path}: {str(e)}")
        return None, None

if __name__ == "__main__":
    print("Secure Encryption System (Text & Images)")
    print("=" * 50)
    
    # Text encryption demo
    text = "This is a secret message!"
    # print("\nText Encryption Demo:")
    print("Original Text:", text)
    
    encrypted_text, iv, keys = encrypt_text(text)
    print("Encrypted Text:", encrypted_text)
    
    decrypted_text = decrypt_text(encrypted_text, iv, keys)
    print("Decrypted Text:", decrypted_text)
    
    # Image encryption demo
    image_path = "input.jpg"  # Change to your image path
    output_dir = "output"
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # print("\nImage Encryption Demo:")
    encrypted_path, decrypted_path = process_image(image_path, output_dir)
    
    if encrypted_path and decrypted_path:
        print("\nOperation successful!")
        print(f"Viewable encrypted image: {encrypted_path}")
        print(f"Decrypted image: {decrypted_path}")
```
